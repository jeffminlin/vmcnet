```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = 20000

energy = np.loadtxt("H2/energy_standard.txt")[:e]
variance = np.loadtxt("H2/variance_standard.txt")[:e]

# energy_20 = np.loadtxt("H2/energy_standard_20.txt")[:e]
# variance_20 = np.loadtxt("H2/variance_standard_20.txt")[:e]

energy_transformed = np.loadtxt("H2/energy_transformed.txt")[:e]
energy_transformed_noclip = np.loadtxt("H2/energy_transformed_noclip.txt")[:e]
variance_transformed = np.loadtxt("H2/variance_transformed.txt")[:e]

# energy_transformed_200k = np.loadtxt("H2/energy_transformed_200k.txt")[:e]
energy_transformed_200k_noclip = np.loadtxt("H2/energy_transformed_200k_noclip.txt")[:e]
# variance_transformed_200k = np.loadtxt("H2/variance_transformed_200k.txt")[:e]

energy_20 = np.loadtxt("H2/energy_standard_20.txt")[:e]
```

```python
ref = energy[-1000:].mean() - 0.05

half_window = 50
window = half_window * 2
epochs = np.arange(half_window, e-half_window + 1)


fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))

# energy_20 = np.convolve(energy_20, np.ones(window)/window, 'valid')
# indices = np.nonzero(energy_20 > ref)
# ax1.plot(epochs[indices], energy_20[indices] -ref, label="Standard, 20 samples", alpha=0.9)

energy = np.convolve(energy, np.ones(window)/window, 'valid')
indices = np.nonzero(energy > ref)
ax1.plot(epochs[indices], energy[indices] -ref, label="Standard, 2k samples")


# energy_transformed = np.convolve(energy_transformed, np.ones(window)/window, 'valid')
# indices = np.nonzero(energy_transformed > ref)
# ax1.plot(epochs[indices], energy_transformed[indices] -ref, label="Transformed, energy_clip", alpha=0.9)

# energy_transformed_noclip = np.convolve(energy_transformed_noclip, np.ones(window)/window, 'valid')
# indices = np.nonzero(energy_transformed_noclip > ref)
# ax1.plot(epochs[indices], energy_transformed_noclip[indices] -ref, label="Transformed, energy_noclip", alpha=0.8)


energy_transformed_noclip = np.convolve(energy_transformed_noclip, np.ones(window)/window, 'valid')
indices = np.nonzero(energy_transformed_noclip > ref)
ax1.plot(epochs[indices], energy_transformed_noclip[indices] -ref, label="Transformed, noclip", alpha=0.8)


energy_transformed_200k_noclip = np.convolve(energy_transformed_200k_noclip, np.ones(window)/window, 'valid')
indices = np.nonzero(energy_transformed_200k_noclip > ref)
ax1.plot(epochs[indices], energy_transformed_200k_noclip[indices] -ref, label="Transformed, 200k samples, energy_noclip", alpha=0.9)

ax1.set_ylim(0, 0.25)
ax1.set_title("Energy")
# ax1.set_yscale('log')
ax1.legend()

ax2.plot(variance, label="Standard")
ax2.plot(variance_transformed, label="Transformed")
# ax2.plot(variance_20, label="Standard, 20 samples", alpha=0.9)
# ax2.plot(variance_transformed_200k, label="Transformed, 200k samples")

ax2.set_title("Local energy variance")
ax2.set_yscale('log')
ax2.legend()

plt.savefig("optimization")
```
